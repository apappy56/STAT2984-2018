```python
# data downloaded from https://dasl.datadescription.com/datafile/couples/

def read_data(filename):
    # read in the file
    f = open(filename,'rU').readlines()
    
    data = []
    for row in f:
        data.append(row.strip().split('\t'))
    return data


filename = 'couples.txt'
# read in the data
data = read_data(filename)

data
```

```python
# convert the list of lists to a dictionary

data_dict = {}
header = data[0]



for row in data[1:]:
    for i,var in enumerate(header):
        if var not in data_dict:
            data_dict[var] = []
        if len(row) == len(header):
            data_dict[var].append([row[i]])

print data_dict.keys()
            
```
